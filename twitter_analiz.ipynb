```python
def timeline_df(tweets):
  df = pd.DataFrame()
  df[ "id" ] = list( map( lambda tweet: tweet.id , tweets ) )
  df[ "olusturulma_zamani" ] = list( map( lambda tweet: tweet.created_at , tweets ) )
  df[ "metin" ] = list( map( lambda tweet: tweet.text , tweets ) )
  df[ "favori_sayisi" ] = list( map( lambda tweet: tweet.favorite_count , tweets ) )
  df[ "retweet_sayisi" ] = list( map( lambda tweet: tweet.retweet_count , tweets ) )
  df[ "kaynak" ] = list( map( lambda tweet: tweet.source , tweets ) )
  df["retweet_durumu"] = list( map( lambda tweet: tweet.retweeted, tweets ) )
  return df
```

```python
def  sentiment_skorla(df):
  text= df["metin"]
  #df["sentiment_skoru"]=[]
  #df["duygu_sinifi"] = []

  for i in range(0,len(text)):
    textB = TextBlob(text[i])
    sentiment_skoru= textB.sentiment.polarity
    df.at[i,"sentiment_skoru"]=sentiment_skoru
    

    if sentiment_skoru < 0.00:
      duygu_sinifi = "Negatif"
      df.at[i,"duygu_sinifi"]=duygu_sinifi
      

    elif sentiment_skoru > 0.00:
      duygu_sinifi = "Pozitif"
      df.at[i,"duygu_sinifi"]=duygu_sinifi

    else:
      duygu_sinifi="Notr"
      df.at[i,"duygu_sinifi"]=duygu_sinifi

  return df
```
